## 팬시인덱싱 전달
기본적인 인덱싱 방법에 추가로 조건에 따른 데이터 샘플링도 가능합니다.  
넘파이의 bool 타입데이터를 인덱스로 전달받는 방법으로 조건에 다른 데이터를 선별한다면 조금 더 고차원적인 데이터 선택이 가능합니다.

```python
df.loc[df['weekday'] == 6, '전력사용량'].mean()
```

```python
# 조건식을 인덱스로 받는 팬시인덱싱
# 건물 번호 기준으로 한 건물에 해당하는 관측치 선별
df.loc[df['num'] == 1, ['전력사용량(kWh)']]
```

```python
# 전력 사용량 기준으로 15000kWh 이상 사용한 관측치 
df.loc[df['전력사용량(kWh)'] >= 15000]
```

```python
# 습도 기준으로 습도가 95이상인 관측치 선별
df.loc[df['습도(%)'] >= 95]
```

```python
# 해당시간 데이터보다 나중에 발생한 관측치 선택 
df.loc[df['date_time'] > '2020-07-31 23']
```

```python
# or 연산 | <--- 양쪽 조건 중 하나라도 참이면 참
# and 연산 & <--- 양쪽 조건이 모두 참이여야 참
# 조건을 2개 사용한 관측치 선택 비전기냉방설비운영, 태양광보유 한 관측치 찾기
# 고온 다습한 기후의 관측치를 선별 기온이 30도 이상이면서 습도도 90 이상인 관측치
df.loc[(df['기온(°C)'] >= 30) & (df['습도(%)'] >= 90)]
```

```python
# 건물번호 기준으로 10번 건물, 20번 건물
df.loc[(df['num'] == 10) | (df['num'] == 20)]
```

```python
# 각각의 값이 명확한 코드를 선호
cond1 = (df['num'] == 10)
cond2 = (df['num'] == 20)
```

```python
df.loc[cond1 | cond2]
```

```python
# 조건식이 아닌 특정 값 기준으로 데이터를 찾을 때
# 필터링 샘플링 하실 때 조건식보다 조금 더 편하실 수 있음
# 건물번호 기준으로 0, 10, 20, 30, 40
df.loc[df['num'].isin([1, 10, 20, 30, 40])]
```

```python
# contains로 전달하는 문자열이 포함되면 True
df.loc[df['date_time'].str.contains('08-01 00')]
```

```python
# 랜덤한 샘플의 수 혹은 비율에 따른 선별 방법
df.sample(frac=0.05)
# n= 데이터 갯수설정
# frac= 데이터 비율설정
```

## 집계값 계산

```python
df.loc[df['전력사용량(kWh)'] == 0]
```

```python
# 전력사용량 기준 집계값 계산
# 평균, 합, 최대값, 최소값, 평균, 분산
print(df['전력사용량(kWh)'].mean())
print(df['전력사용량(kWh)'].sum())
print(df['전력사용량(kWh)'].max())
print(df['전력사용량(kWh)'].min())
print(df['전력사용량(kWh)'].std()) # 표준편차
print(df['전력사용량(kWh)'].var()) # 분산
print(df['전력사용량(kWh)'].argmax()) # 최대값 인덱스를 출력
print(df['전력사용량(kWh)'].argmin()) # 최대값 인덱스를 출력
```

```python
# 변수데이터의 분포가 실수형 데이터 연속형 분포를 가진 데이터에 적용
# 컬럼 값 기준 가장 큰, 작은 값을 가진 샘플 확인 nlargest
df['전력사용량(kWh)'].nlargest(10)
# 컬럼 값 기준으로 가장 작은 값 nsmallest
df['전력사용량(kWh)'].nsmallest(10)
```

```python
print(df['태양광보유'].unique()) # 중복값을 제외한 고윳값을 출력
print(df['태양광보유'].nunique()) # 고윳값 갯수 출력
df['전력사용량(kWh)'].value_counts() # 변수의 고윳값을 카운트 해서 내림차순으로 정렬
```

## 데이터 재구조화
기존의 데이터 샘플링과는 달리 기준점으로 생각할 수 있는 컬럼 값을 기준으로 새롭게 데이터 프레임을 생성하며  
평균, 합, 카운트 등을 통해 데이터를 다차원적으로 분석 할 수 있는 함수를 제공합니다.
> 기준 변수(컬럼)가 한개 일 경우
>> **`df`**.**`groupby('컬럼명')`**.**`agg(집계방법)`**  

> 기준 변수(컬럼)가 두개 이상일 경우  
>>  **`pd`**.**`pivot_table(data=데이터프레임명, index=기준컬럼1, columns=기준컬럼2, values=집계데이터, aggfunc=집계방법)`**  

>> 적용가능한 통계 함수

| 함수명 | 내용 |
|-|-|
| count | 갯수 |
| sum | 합 |
| mean | 평균 |
| median | 중앙값 |
| var, std | 분산, 표준편차 |
| min, max | 최소값, 최대값 |
| unique, nunique | 고윳값, 고윳값 갯수 |
| prod | 곱 |

### groupby

```python
df.head()
```

```python
# groupby 변수의 분포가 카테고리컬 분포를 갖는 데이터가 좋습니다.
df.groupby('num', as_index=False).mean(numeric_only=True)
# as_index=False 인덱스 초기화
# numeric_only=True 숫자형 데이터만 집계값 계산 함수를 적용
```

```python
df.head()
```

```python
# 전력사용량, 최대값, 평균
df.groupby('num').agg({'전력사용량':['max', 'mean'],
                       '기온':'mean'})
```

### pivot_table

```python
df.head()
```

## 데이터프레임 병합
> 실제 분석업무를 진행하다보면 데이터가 여기저기 분산되어 있을 경우가 더 많습니다.  
조각난 데이터를 분석에 필요한 데이터셋으로 만들기 위해 데이터프레임 병합을 많이 사용합니다.  
한개 이상의 데이터프레임을 병합 할 때 주로 사용하는 함수 2가지를 알아보겠습니다.    

### 데이터 병합에 사용가능한 key(병합할 기준이 되는 행 or 열)값이 있는경우
> **`pd`**.**`merge(베이스데이터프레임, 병합할데이터프레임)`**

>> 사용가능한 파라메터
>> - how : 'left', 'right', 'inner', 'outer'
>> - left_on : key값이 다를 경우 베이스데이터프레임의 key 설정
>> - right_on : key값이 다를 경우 병합데이터프레임의 key 설정    


```python
merge_df1 = pd.DataFrame({
    '이름': ['원영', '사쿠라', '유리', '예나', '유진', '나코', '은비', '혜원', '히토미', '채원', '민주', '째욘'],
    '국어': [100, 70, 70, 70, 60, 90, 90, 70, 70, 80, 100, 100],
    '영어': [100, 90, 80, 50, 70, 100, 70, 90, 100, 100, 80, 100]
    }, columns=['이름', '국어', '영어'])

merge_df2 = pd.DataFrame({
    '일어': [80, 100, 100, 90, 70, 50, 100],
    '수학': [90, 70, 100, 80, 70, 80, 90],
    '이름': ['원영', '사쿠라', '나코', '히토미', '예나', '은비', '째욘'],
    }, columns=['일어', '수학', '이름'])
```

```python
merge_df2
```

```python
# merge 테스트
pd.merge(merge_df1, merge_df2, how='outer')
# 양쪽 데이터프레임에 key값에 해당하는 데이터가 모두 있는 경우 병합 'inner' and 조건으로 묶임
# 한쪽 데이터프레임에만 key값 데이터가 존재해도 데이터를 병합 'outer' or 조건으로 병합
# 'left' 처음 전달하는 데이터프레임 기준으로 병합, 두번째 전달 데이터프레임 기준 병합 'right'
```

### apply 함수로 컬럼에 함수 적용
> 인스턴스 함수인 lambda 와 apply 함수를 사용하여 인자로 받는 모든 데이터에 함수를 적용  
커스텀 함수 적용도 가능

> apply 함수로 컬럼에 적용시키는 코드 구조  
 **`df['컬럼명']`** = **`df['컬럼명']`**.**`apply(함수명)`**

```python
# 함수
# y = f(x)
# 기본적인 함수의 구조
def make_f(x): # make_f라는 이름의 기능명(함수명)을 선언 x라는 입력값도 동시에 설정
    return (x * 1.8) + 32 # 함수를 호출하여 실행할 때 실행시키는 실행코드, return을 통해 결과값을 생성
```

```python
make_f(28)
```

```python
# 화씨 컬럼 추가 (℃ × 1.8) ＋ 32
df['화씨'] = df['기온'].apply(make_f)
```

```python
# 조건문 if, elif, else
# 제어구문 코드의 실행이 이루어지는 조건은 조건 하나만 실행
if True: # 만약 조건식이 참이라면 아래 실행코드가 실행
    print('참입니다.')
elif True: # 만약 첫번째 조건이 아닌 다른 조건이 참이라면
    print('두번 째 참입니다.')
else: # 위 모든 조건이 거짓이면
    print('모두 거짓입니다.')
```

```python
# 복잡한 형태의 함수를 제작
# 12시 ~ 16시 True, 그 외 시간대는 False 피크시간대라는 이름으로 변수를 생성
def make_peak(x):
    if 12 <= x < 17:
        return True
    else:
        return False
```
