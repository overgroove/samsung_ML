## 팬시인덱싱 전달
기본적인 인덱싱 방법에 추가로 조건에 따른 데이터 샘플링도 가능합니다.  
넘파이의 bool 타입데이터를 인덱스로 전달받는 방법으로 조건에 다른 데이터를 선별한다면 조금 더 고차원적인 데이터 선택이 가능합니다.

```python
# 조건식을 인덱스로 받는 팬시인덱싱
# 건물 번호 1 기준으로 한 건물에 해당하는 관측치 선별
df.loc[df['num'] == 1]
```

```python
# 전력 사용량 기준으로 15000kWh 이상 사용한 관측치 
df.loc[df['전력사용량(kWh)'] >= 15000]
```

```python
# 해당시간 데이터보다 나중에 발생한 관측치 선택 
df.loc[df['date_time'] > '2020-07-31 23']
```

```python
# or 연산 | 
# and 연산 &
# 조건을 2개 사용한 관측치 선택 비전기냉방설비운영, 태양광보유 한 관측치 찾기
# 고온다습한 기후 데이터를 선별 기온 30도 이상, 습도도 90 이상인 관측치 선별
df.loc[(df['기온(°C)'] >= 30) & (df['습도(%)'] >= 90)]
```

```python
# 건물 번호가 10번 건물데이터와 20번 건물 데이터
df.loc[(df['num'] == 10) | (df['num'] == 20)]
```

```python
cond1 = (df['num'] == 10)
cond2 = (df['num'] == 20)
df.loc[cond1|cond2]
```

```python
# 조건식이 아닌 특정 값 기준으로 데이터를 찾을 때
# 필터링 샘플링 하실 때 조건식보다 조금 더 편하실 수 있음
df.loc[df['num'].isin([1, 2, 3, 4, 5])]
# isin([리스트형태로 전달하는 데이터에 포함이 된다면 True])
# 찾아야 하는 값을 정확하게 알고 있을 때 사용 가능
```

```python
# contains로 전달하는 문자열이 포함되면 True
df.loc[df['date_time'].str.contains('08-01 23')]
```

```python
# 랜덤한 샘플의 수 혹은 비율에 따른 선별 방법
df.sample(frac=0.05)
# n=데이터 갯수 지정, frac=비율지정
```

## 데이터 재구조화
기존의 데이터 샘플링과는 달리 기준점으로 생각할 수 있는 컬럼 값을 기준으로 새롭게 데이터 프레임을 생성하며  
평균, 합, 카운트 등을 통해 데이터를 다차원적으로 분석 할 수 있는 함수를 제공합니다.
> 기준 변수(컬럼)가 한개 일 경우
>> **`df`**.**`groupby('컬럼명')`**.**`agg(집계방법)`**  

> 기준 변수(컬럼)가 두개 이상일 경우  
>>  **`pd`**.**`pivot_table(data=데이터프레임명, index=기준컬럼1, columns=기준컬럼2, values=집계데이터, aggfunc=집계방법)`**  

>> 적용가능한 통계 함수

| 함수명 | 내용 |
|-|-|
| count | 갯수 |
| sum | 합 |
| mean | 평균 |
| median | 중앙값 |
| var, std | 분산, 표준편차 |
| min, max | 최소값, 최대값 |
| unique, nunique | 고윳값, 고윳값 갯수 |
| prod | 곱 |

### groupby

```python
df.head()
```

```python
# groupby 기준변수가 한개 일 경우 추천 (여러개도 가능) - 변수값의 고윳값에 따라 데이터 계산이 이루어지기에 카테고리컬 데이터 좋습니다.
# num 건물별로 생각
df.groupby('num').mean(numeric_only=True)
df.groupby('num', as_index=False).agg({'전력사용량(kWh)':['mean', 'min'],
                                       '기온(°C)':['max', 'var']})
```

### pivot_table

```python
df.head()
```

```python
# pivot_table
# 그룹화 시켜 생각해야 할 변수가 2개일 경우 ---> 2개의 변수가 카테고리면 좋음
df.pivot_table(index='태양광보유',
               columns='비전기냉방설비운영',
               values='전력사용량(kWh)',
               aggfunc='count')
# 코드자체는 groupby보다 간편 --> 
```

>> - how : 'left', 'right', 'inner', 'outer'
>> - left_on : key값이 다를 경우 베이스데이터프레임의 key 설정
>> - right_on : key값이 다를 경우 병합데이터프레임의 key 설정
