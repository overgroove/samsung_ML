## 팬시인덱싱 전달
기본적인 인덱싱 방법에 추가로 조건에 따른 데이터 샘플링도 가능합니다.  
넘파이의 bool 타입데이터를 인덱스로 전달받는 방법으로 조건에 다른 데이터를 선별한다면 조금 더 고차원적인 데이터 선택이 가능합니다.

```python
df.loc[df['강수량(mm)'] >= 20, ['기온(°C)','풍속(m/s)','습도(%)']]
```

```python
# 조건식을 인덱스로 받는 팬시인덱싱
# 건물 번호 기준으로 한 건물에 해당하는 관측치 선별
# 조건식을 전달하고 True인 관측치만 행기준으로 가져온다
df.loc[df['num'] == 1]
```

```python
# 전력 사용량 기준으로 15000kWh 이상 사용한 관측치 
df.loc[df['전력사용량(kWh)'] >= 15000]
```

```python
# 습도가 100인 관측치 선별
df.loc[df['습도(%)'] > 99.99]
```

```python
# 해당시간 데이터보다 나중에 발생한 관측치 선택 
df.loc[df['date_time'] > '2020-08-20 00']
```

```python
# or 연산 | 한쪽만 충족이 되어도 참
# and 연산 & 두 조건이 모두 만족이 되어야 참
# 조건을 2개 사용한 관측치 선택 비전기냉방설비운영, 태양광보유 한 관측치 찾기
# 고온 다습한 관측치 온도가 30도 이상, 습도가 90 이상
cond1 = (df['기온(°C)'] >= 30)
cond2 = (df['습도(%)'] >= 90)
df.loc[cond1 & cond2]
```

```python
# 건물번호가 10번 20번인 관측치 선택
df.loc[(df['num'] == 10) | (df['num'] == 20)]
```

```python
# 조건식이 아닌 특정 값 기준으로 데이터를 찾을 때
# 필터링 샘플링 하실 때 조건식보다 조금 더 편하실 수 있음
df.loc[df['num'].isin([1, 5, 7, 9, 60])]
# 데이터의 값을 정확하게 알고 있을 때 조건에 따른 데이터 선별보다 편한 방식
```

```python
# contains로 전달하는 문자열이 포함되면 True
df.loc[df['date_time'].str.contains(' 00')]
```

```python
# 랜덤한 샘플의 수 혹은 비율에 따른 선별 방법
# 조건에 의한 데이터 선택 X, 랜덤한 방식으로 데이터 선별
df.sample(frac=0.2)
# n 전달한 데이터 갯수만큼 가져와라
# frac 비율전달
```

## 집계값 계산

```python
# 전력사용량 기준 집계값 계산
# 평균, 합, 최대값, 최소값, 평균, 분산
print(df['전력사용량(kWh)'].mean())
print(df['전력사용량(kWh)'].sum())
print(df['전력사용량(kWh)'].max())
print(df['전력사용량(kWh)'].min())
print(df['전력사용량(kWh)'].std()) # 표준편차
print(df['전력사용량(kWh)'].var()) # 분산
print(df['전력사용량(kWh)'].argmax()) # 최대값이 있는 인덱스
print(df['전력사용량(kWh)'].argmin()) # 최소값이 있는 인덱스
print(df['전력사용량(kWh)'].unique()) # 고윳값 출력
print(df['전력사용량(kWh)'].nunique())
print(df['전력사용량(kWh)'].value_counts()) # 고윳값의 카운트 내림차순 정렬
```

```python
# 카테고리컬 데이터의 대소비교 큰 값을 찾는 용도로 많이 사용합니다.
df['태양광보유'].value_counts()
```

## 데이터 재구조화
기존의 데이터 샘플링과는 달리 기준점으로 생각할 수 있는 컬럼 값을 기준으로 새롭게 데이터 프레임을 생성하며  
평균, 합, 카운트 등을 통해 데이터를 다차원적으로 분석 할 수 있는 함수를 제공합니다.
> 기준 변수(컬럼)가 한개 일 경우
>> **`df`**.**`groupby('컬럼명')`**.**`agg(집계방법)`**  

> 기준 변수(컬럼)가 두개 이상일 경우  
>>  **`pd`**.**`pivot_table(data=데이터프레임명, index=기준컬럼1, columns=기준컬럼2, values=집계데이터, aggfunc=집계방법)`**  

>> 적용가능한 통계 함수

| 함수명 | 내용 |
|-|-|
| count | 갯수 |
| sum | 합 |
| mean | 평균 |
| median | 중앙값 |
| var, std | 분산, 표준편차 |
| min, max | 최소값, 최대값 |
| unique, nunique | 고윳값, 고윳값 갯수 |
| prod | 곱 |

## 데이터프레임 병합
> 실제 분석업무를 진행하다보면 데이터가 여기저기 분산되어 있을 경우가 더 많습니다.  
조각난 데이터를 분석에 필요한 데이터셋으로 만들기 위해 데이터프레임 병합을 많이 사용합니다.  
한개 이상의 데이터프레임을 병합 할 때 주로 사용하는 함수 2가지를 알아보겠습니다.    

### 데이터 병합에 사용가능한 key(병합할 기준이 되는 행 or 열)값이 있는경우
> **`pd`**.**`merge(베이스데이터프레임, 병합할데이터프레임)`**

>> 사용가능한 파라메터
>> - how : 'left', 'right', 'inner', 'outer'
>> - left_on : key값이 다를 경우 베이스데이터프레임의 key 설정
>> - right_on : key값이 다를 경우 병합데이터프레임의 key 설정    


```python
merge_df1 = pd.DataFrame({
    '이름': ['원영', '사쿠라', '유리', '예나', '유진', '나코', '은비', '혜원', '히토미', '채원', '민주', '째욘'],
    '국어': [100, 70, 70, 70, 60, 90, 90, 70, 70, 80, 100, 100],
    '영어': [100, 90, 80, 50, 70, 100, 70, 90, 100, 100, 80, 100]
    }, columns=['이름', '국어', '영어'])

merge_df2 = pd.DataFrame({
    '일어': [80, 100, 100, 90, 70, 50, 100],
    '수학': [90, 70, 100, 80, 70, 80, 90],
    'name': ['원영', '사쿠라', '나코', '히토미', '예나', '은비', '째욘'],
    }, columns=['일어', '수학', 'name'])
```

```python
merge_df2
```

```python
# merge 테스트
# 교집합, 동시충족, and, inner
# 합집합, 조건충족, or, outer
pd.merge(merge_df1, merge_df2, how='inner', left_on='이름', right_on='name')
# key 값이 없는경우 병합을 위해 사용할 key 변수명을 강제지정 할 수 있는 파라메터
```

### 결측치 비율 계산

```python
# 컬럼명을 순환하면서 결측치 비율 계산
for col_nm in df.columns:
    print(col_nm, df[col_nm].isna().sum() / len(df) * 100)
```

```python
# 분석가 기준으로 삼는 비율이 다 다름 도메인따라 달라지기도 함.
```

```python
# fillna() 함수로 NaN 값을 컬럼의 평균으로 채우기
df['기온(°C)'].fillna(df['기온(°C)'].mean(), inplace=True) # 전달하는 값으로 결측데이터를 치환(대체)
# 기온 변수는 실수형 데이터, 연속형데이터 ---> 평균값 채워넣는 작업
# 범주형(카테고리컬) 데이터, 최빈값
```

### 결측치 제거

```python
# 마지막에 합니다.
# 행기준 데이터 혹은 열기준으로 결측데이터가 하나라도 있으면 해당 행, 열을 전체 삭제하는 방법으로 결측치를 처리
# 결측데이터를 모두 채워넣는 작업을 완료 하고 가장 마지막에 합니다. ---> 데이터 버림(수집이 어려움)
df.dropna() # 행기준 axis=1
```
