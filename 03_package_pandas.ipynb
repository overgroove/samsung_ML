## 팬시인덱싱 전달
기본적인 인덱싱 방법에 추가로 조건에 따른 데이터 샘플링도 가능합니다.  
넘파이의 bool 타입데이터를 인덱스로 전달받는 방법으로 조건에 다른 데이터를 선별한다면 조금 더 고차원적인 데이터 선택이 가능합니다.

```python
df[df['num'] > 55] # 뷰값만 확인 ---> 결과값의 원본값 변경 불가능
df.loc[df['num'] > 55] # 메모리 상 값에 접근 ---> 원본값 변경 가능
```

```python
# 조건식을 인덱스로 받는 팬시인덱싱
# 건물 번호 기준으로 한 건물에 해당하는 관측치 선별
df.loc[df['num'] == 1]
# 엑셀 필터적용
```

```python
# 전력 사용량 기준으로 15000kWh 이상 사용한 관측치 
df.loc[df['전력사용량(kWh)'] > 15000]
```

```python
# 해당시간 데이터보다 나중에 발생한 관측치 선택 
df.loc[df['date_time'] > '2020-07-31 23']
```

```python
# or 연산 | 
# and 연산 &
# 조건을 2개 사용한 관측치 선택 고온, 다습했던 관측치 선별 고온 30도 이상, 다습 습도 90 이상
df.loc[(df['기온(°C)'] >= 30) & (df['습도(%)'] >= 90)]
```

```python
# 건물번호 기준으로 15, 45 두개 가져온다
df.loc[(df['num'] == 15) | (df['num'] == 45)]
```

```python
# 각 조건을 변수 형태로 저장 후 loc 내부에서 &, | 연산식 계산
cond1 = (df['num'] == 15)
cond2 = (df['num'] == 45)
df.loc[cond1 | cond2]
```

```python
# 조건식이 아닌 특정 값 기준으로 데이터를 찾을 때
# 필터링 샘플링 하실 때 조건식보다 조금 더 편하실 수 있음
df.loc[df['num'].isin([10, 20, 30, 40, 50, 60])]
# 변수 내 데이터의 정확한 값을 알고 있을 때 사용 가능
```

```python
# contains로 전달하는 문자열이 포함되면 True
df.loc[df['date_time'].str.contains('08-09')]
```

```python
# 랜덤한 샘플의 수 혹은 비율에 따른 선별 방법
df.sample(frac=0.1)
# n=관측치 갯수 설정, frac=데이터 비율 설정
```

## 집계값 계산

```python
# 전력사용량 기준 집계값 계산
# 평균, 중앙값, 합, 최대값, 최소값, 표준편차, 분산
print(df['전력사용량(kWh)'].mean())
print(df['전력사용량(kWh)'].median())
print(df['전력사용량(kWh)'].sum())
print(df['전력사용량(kWh)'].max())
print(df['전력사용량(kWh)'].min())
print(df['전력사용량(kWh)'].std())
print(df['전력사용량(kWh)'].var())
print(df['전력사용량(kWh)'].argmax()) # 최대값의 인덱스를 리턴
print(df['전력사용량(kWh)'].argmin()) # 최소값의 인덱스를 리턴
```

## 데이터 재구조화
기존의 데이터 샘플링과는 달리 기준점으로 생각할 수 있는 컬럼 값을 기준으로 새롭게 데이터 프레임을 생성하며  
평균, 합, 카운트 등을 통해 데이터를 다차원적으로 분석 할 수 있는 함수를 제공합니다.
> 기준 변수(컬럼)가 한개 일 경우
>> **`df`**.**`groupby('컬럼명')`**.**`agg(집계방법)`**  

> 기준 변수(컬럼)가 두개 이상일 경우  
>>  **`pd`**.**`pivot_table(data=데이터프레임명, index=기준컬럼1, columns=기준컬럼2, values=집계데이터, aggfunc=집계방법)`**  

>> 적용가능한 통계 함수

| 함수명 | 내용 |
|-|-|
| count | 갯수 |
| sum | 합 |
| mean | 평균 |
| median | 중앙값 |
| var, std | 분산, 표준편차 |
| min, max | 최소값, 최대값 |
| unique, nunique | 고윳값, 고윳값 갯수 |
| prod | 곱 |

### groupby

```python
# groupby 대상 변수가 카테고리컬 데이터
df.groupby('num', as_index=False).mean(numeric_only=True)
```

```python
df.groupby('num').agg({'전력사용량(kWh)':['mean', 'max', 'min'],
                       '기온(°C)':['max', 'var']}, 
                       numeric_only=True)
```

### pivot_table

## 데이터프레임 병합
> 실제 분석업무를 진행하다보면 데이터가 여기저기 분산되어 있을 경우가 더 많습니다.  
조각난 데이터를 분석에 필요한 데이터셋으로 만들기 위해 데이터프레임 병합을 많이 사용합니다.  
한개 이상의 데이터프레임을 병합 할 때 주로 사용하는 함수 2가지를 알아보겠습니다.    

### 데이터 병합에 사용가능한 key(병합할 기준이 되는 행 or 열)값이 있는경우
> **`pd`**.**`merge(베이스데이터프레임, 병합할데이터프레임)`**

>> 사용가능한 파라메터
>> - how : 'left', 'right', 'inner', 'outer'
>> - left_on : key값이 다를 경우 베이스데이터프레임의 key 설정
>> - right_on : key값이 다를 경우 병합데이터프레임의 key 설정    


```python
merge_df1 = pd.DataFrame({
    '이름': ['원영', '사쿠라', '유리', '예나', '유진', '나코', '은비', '혜원', '히토미', '채원', '민주', '째욘'],
    '국어': [100, 70, 70, 70, 60, 90, 90, 70, 70, 80, 100, 100],
    '영어': [100, 90, 80, 50, 70, 100, 70, 90, 100, 100, 80, 100]
    }, columns=['이름', '국어', '영어'])

merge_df2 = pd.DataFrame({
    '일어': [80, 100, 100, 90, 70, 50, 100],
    '수학': [90, 70, 100, 80, 70, 80, 90],
    'name': ['원영', '사쿠라', '나코', '히토미', '예나', '은비', '째욘'],
    }, columns=['일어', '수학', 'name'])
```

```python
merge_df2
```

```python
# merge 테스트
pd.merge(merge_df1, merge_df2, left_on='이름', right_on='name', how='left')
# how 전달하는 방법에 따라 병합 방식이 달라집니다.
# inner 양쪽 데이터프레임에 데이터가 key값의 데이터가 모두 있는 경우 and
# outer 양쪽 데이터프레임 중 한쪽에만 key값 데이터가 있어도 병합 or
# left 먼저 전달하는 데이터프레임을 베이스데이터프레임으로 병합
# right 뒤에 전달하는 데이터프레임을 베이스데이터프레임으로 병합
# left_on, right_on key값의 변수명이 다를 경우 수동으로 key값을 지정하는 옵션
```
