## 팬시인덱싱 전달
기본적인 인덱싱 방법에 추가로 조건에 따른 데이터 샘플링도 가능합니다.  
넘파이의 bool 타입데이터를 인덱스로 전달받는 방법으로 조건에 다른 데이터를 선별한다면 조금 더 고차원적인 데이터 선택이 가능합니다.

```python
# 조건식을 인덱스로 받는 팬시인덱싱
# 건물 번호 1 기준으로 한 건물에 해당하는 관측치 선별
df.loc[df['num'] == 1]
```

```python
# 전력 사용량 기준으로 15000kWh 이상 사용한 관측치 
df.loc[df['전력사용량(kWh)'] >= 15000]
```

```python
# 해당시간 데이터보다 나중에 발생한 관측치 선택 
df.loc[df['date_time'] > '2020-07-31 23']
```

```python
# or 연산 | 
# and 연산 &
# 조건을 2개 사용한 관측치 선택 비전기냉방설비운영, 태양광보유 한 관측치 찾기
# 고온다습한 기후 데이터를 선별 기온 30도 이상, 습도도 90 이상인 관측치 선별
df.loc[(df['기온(°C)'] >= 30) & (df['습도(%)'] >= 90)]
```

```python
# 건물 번호가 10번 건물데이터와 20번 건물 데이터
df.loc[(df['num'] == 10) | (df['num'] == 20)]
```

```python
cond1 = (df['num'] == 10)
cond2 = (df['num'] == 20)
df.loc[cond1|cond2]
```

```python
# 조건식이 아닌 특정 값 기준으로 데이터를 찾을 때
# 필터링 샘플링 하실 때 조건식보다 조금 더 편하실 수 있음
df.loc[df['num'].isin([1, 2, 3, 4, 5])]
# isin([리스트형태로 전달하는 데이터에 포함이 된다면 True])
# 찾아야 하는 값을 정확하게 알고 있을 때 사용 가능
```

```python
# contains로 전달하는 문자열이 포함되면 True
df.loc[df['date_time'].str.contains('08-01 23')]
```

```python
# 랜덤한 샘플의 수 혹은 비율에 따른 선별 방법
df.sample(frac=0.05)
# n=데이터 갯수 지정, frac=비율지정
```

## 데이터 재구조화
기존의 데이터 샘플링과는 달리 기준점으로 생각할 수 있는 컬럼 값을 기준으로 새롭게 데이터 프레임을 생성하며  
평균, 합, 카운트 등을 통해 데이터를 다차원적으로 분석 할 수 있는 함수를 제공합니다.
> 기준 변수(컬럼)가 한개 일 경우
>> **`df`**.**`groupby('컬럼명')`**.**`agg(집계방법)`**  

> 기준 변수(컬럼)가 두개 이상일 경우  
>>  **`pd`**.**`pivot_table(data=데이터프레임명, index=기준컬럼1, columns=기준컬럼2, values=집계데이터, aggfunc=집계방법)`**  

>> 적용가능한 통계 함수

| 함수명 | 내용 |
|-|-|
| count | 갯수 |
| sum | 합 |
| mean | 평균 |
| median | 중앙값 |
| var, std | 분산, 표준편차 |
| min, max | 최소값, 최대값 |
| unique, nunique | 고윳값, 고윳값 갯수 |
| prod | 곱 |

### groupby

```python
df.head()
```

```python
# groupby 기준변수가 한개 일 경우 추천 (여러개도 가능) - 변수값의 고윳값에 따라 데이터 계산이 이루어지기에 카테고리컬 데이터 좋습니다.
# num 건물별로 생각
df.groupby('num').mean(numeric_only=True)
df.groupby('num', as_index=False).agg({'전력사용량(kWh)':['mean', 'min'],
                                       '기온(°C)':['max', 'var']})
```

### pivot_table

```python
df.head()
```

## 데이터프레임 병합
> 실제 분석업무를 진행하다보면 데이터가 여기저기 분산되어 있을 경우가 더 많습니다.  
조각난 데이터를 분석에 필요한 데이터셋으로 만들기 위해 데이터프레임 병합을 많이 사용합니다.  
한개 이상의 데이터프레임을 병합 할 때 주로 사용하는 함수 2가지를 알아보겠습니다.    

### 데이터 병합에 사용가능한 key(병합할 기준이 되는 행 or 열)값이 있는경우
> **`pd`**.**`merge(베이스데이터프레임, 병합할데이터프레임)`**

>> 사용가능한 파라메터
>> - how : 'left', 'right', 'inner', 'outer'
>> - left_on : key값이 다를 경우 베이스데이터프레임의 key 설정
>> - right_on : key값이 다를 경우 병합데이터프레임의 key 설정    


```python
merge_df1 = pd.DataFrame({
    '이름': ['원영', '사쿠라', '유리', '예나', '유진', '나코', '은비', '혜원', '히토미', '채원', '민주', '째욘'],
    '국어': [100, 70, 70, 70, 60, 90, 90, 70, 70, 80, 100, 100],
    '영어': [100, 90, 80, 50, 70, 100, 70, 90, 100, 100, 80, 100]
    }, columns=['이름', '국어', '영어'])

merge_df2 = pd.DataFrame({
    '일어': [80, 100, 100, 90, 70, 50, 100],
    '수학': [90, 70, 100, 80, 70, 80, 90],
    '이름': ['원영', '사쿠라', '나코', '히토미', '예나', '은비', '째욘'],
    }, columns=['일어', '수학', '이름'])
```

```python
merge_df2
```

```python
# merge 테스트
merge_df = pd.merge(merge_df1, merge_df2, how='right')
# 기본설정은 양쪽 데이터프레임의 key값 변수의 데이터가 모두 있는 경우에 데이터프레임 병합을 수행합니다. how='inner' and
# key값 데이터가 한쪽 데이터프레임에만 존재해도 데이터프레임을 병합 or
# 먼저 전달하는 데이터프레임을 기준으로 병합 how='left'
# 이후에 전달하는 데이터프레임을 기준으로 병합 how='right'
```

```python
merge_df
```

### apply 함수로 컬럼에 함수 적용

 **`df['컬럼명']`** = **`df['컬럼명']`**.**`apply(함수명)`**

```python
# 함수 function
# y = f(x)
# 기본적인 함수의 구조
def make_sqaure(x): # make_square이름값을 앞으로 내가 사용할 기능명(함수명으로) 선언하면서 입력값으로 x이름의 파라메터를 설정
    return x ** 2
```

## 결측치 처리
> 데이터 분석을 위해서는 데이터셋 내에 빈 값이 있는 경우 분석에 방해가 될 수 있는 여지가 많습니다.

### 결측치 비율 계산

```python
# 결측치 갯수 세기
df['기온'].isna().sum()
# 비율계산
df['기온'].isna().sum() / len(df) * 100
```

```python
# 컬럼명을 순환하면서 결측치 비율 계산
for col_nm in df.columns:
    print(col_nm, df[col_nm].isna().sum() / len(df) * 100)
```

```python
# 연속형 데이터의 결측치는 평균(중앙값)
# fillna() 함수로 NaN 값을 컬럼의 평균으로 채우기
df['기온'].fillna(df['기온'].mean(), inplace=True)
```

```python
# 카테고리컬 데이터의 결측처리 --> 최빈값
df['비전기냉방설비운영'].fillna(1, inplace=True)
```

```python
df['비전기냉방설비운영'].mode()[0] # 최빈값 
```

```python
# 건물별 평균값으로 결측치 처리
for i in range(1, 61):
    df.loc[(df['num'] == i) & (df['풍속'].isna()), '풍속'] = df.loc[df['num'] == i, '풍속'].mean()
```

```python
for col_nm in df.columns[5:12]:
    if df[col_nm].nunique() < 5: # 카테고리컬 데이터 처리 조건
        df[col_nm].fillna(df[col_nm].mode()[0], inplace=True)
        print(f'{col_nm} 결측 처리 완료!')
    else:
        df[col_nm].fillna(df[col_nm].mean(), inplace=True)
        print(f'{col_nm} 결측 처리 완료!')
```
