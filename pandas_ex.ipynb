**구매자의 사용언어에 따라 총 구매금액을 구하시오**

```python
df.groupby('Language').sum()['Purchase Price'].sort_values(ascending=False)
```

```python
df.groupby('Language').agg({'Purchase Price':'sum'})
```

**오전과 오후에 주문한 총 구매금액을 구하시오**

```python
# CODE HERE
df.groupby('AM or PM').agg({'Purchase Price':['sum', 'max']})
```

**평균 구매금액이 큰 상위 10개 직업을 출력하시오**

```python
df.groupby('Job').mean()['Purchase Price'].sort_values(ascending=False)
```

```python
# CODE HERE
df.groupby('Job').agg({'Purchase Price':'mean'}).nlargest(10, 'Purchase Price') # .sort_values(by='Purchase Price')
```

**카드사별 오전과 오후 평균 구매금액 구하시오**

```python
# CODE HERE
pd.pivot_table(data=df,
               index='CC Provider',
               columns='AM or PM',
               values='Purchase Price',
               aggfunc='mean')
```

**아파트 거주 고객의 직업별 총 구매금액**

```python
# CODE HERE
df.loc[df['Address'].str.contains('Apt.')].groupby('Job').agg({'Purchase Price':'sum'})
```

```python
df['Address'].str.contains('Apt.')
```

```python
pd.pivot_table(data=df,
               index='Job',
               columns=df['Address'].str.contains('Apt.'),
               values='Purchase Price',
               aggfunc='sum')
```
