**구매자의 사용언어에 따라 총 구매금액을 구하시오**

```python
# CODE HERE
df.groupby('Language').sum(numeric_only=True)['Purchase Price']
df.groupby('Language').agg({'Purchase Price':'sum'})
```

**오전과 오후에 주문한 총 구매금액을 구하시오**

```python
# CODE HERE
df.groupby('AM or PM').sum(numeric_only=True)['Purchase Price']
df.groupby('AM or PM').agg({'Purchase Price':'sum'})
```

**평균 구매금액이 큰 상위 10개 직업을 출력하시오**

```python
# CODE HERE
df.groupby('Job').mean(numeric_only=True)['Purchase Price'].nlargest(10)
df.groupby('Job').agg({'Purchase Price':'mean'}).nlargest(10, 'Purchase Price')
```

**카드사별 오전과 오후 평균 구매금액 구하시오**

```python
# CODE HERE
df.pivot_table(index='CC Provider',
               columns='AM or PM',
               values='Purchase Price',
               aggfunc=['mean', 'sum'])
```

**아파트 거주 고객의 직업별 총 구매금액**

```python
# CODE HERE
df.loc[df['Address'].str.contains('Apt.')].groupby('Job').sum(numeric_only=True)['Purchase Price']
df.loc[df['Address'].str.contains('Apt.')].groupby('Job').agg({'Purchase Price':'sum'})
```
